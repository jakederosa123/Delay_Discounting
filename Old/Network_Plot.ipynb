```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import spearmanr, pearsonr
from sklearn.model_selection import train_test_split
from time import time
import pandas as pd 
import numpy as np
from scipy import stats
import nilearn
import nibabel as nib
from nilearn.input_data import NiftiLabelsMasker
import os
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
from igraph import *
from matplotlib.colors import LinearSegmentedColormap
import community as community_louvain
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import networkx as nx
```

```python
X = pd.read_csv('C:/Users/jacob.derosa/Desktop/Scripts/Temp_Disc_Splits/Subtype_Data/X_Stab.csv',  index_col=0)
G = nx.from_pandas_adjacency(X)
threshold = 0.6
G.remove_edges_from([(n1, n2) for n1, n2, w in G.edges(data="weight") if w < threshold])
pos = nx.spring_layout(G)
graph = Graph.Weighted_Adjacency(X.values.tolist(), mode=ADJ_UNDIRECTED, attr="weight")
Louvain = graph.community_multilevel(weights=graph.es['weight'])
cmap = LinearSegmentedColormap.from_list('mycmap', ["#6faac9", "#7a5195", "#ef5675", "#ffa600"])
nx.draw_networkx_nodes(G, pos, node_size=150,
                       cmap=cmap, node_color=list(Louvain.membership),
                      edgecolors = "black")
nx.draw_networkx_edges(G, pos, alpha=1)
fig = plt.gcf()
fig.savefig('CBCL_Network.png', dpi=300)
plt.show()
```
